```python
#frequency of words
fd = nltk.FreqDist(bible)
```

```python
fd.most_common(20)
```

```python
sw = set(nltk.corpus.stopwords.words ('english'))
print(len(sw))
list(sw) [:10]
```

```python
#use the stopwords to filter the file in the macbeth variable. there are 179 stopwords in english language
bible_filtered = [w for w in bible if w.lower() not in sw]
fd = nltk.FreqDist (bible_filtered)
fd.most_common(10)
```

```python
#some punctuations are still popping up, romove them
import string
punctuation = set (string.punctuation)
bible_filtered2 = [w.lower () for w in bible if w.lower () not in sw and w.lower () not in punctuation]
```

```python
fd = nltk.FreqDist (bible_filtered2)
fd.most_common(10)
```

```python
# SELECTIon of words from texts
# selecting them based on their lengths
long_words = [w for w in bible if len(w)> 5]
sorted(long_words)
```

```python
faith_words = [w for w in bible if 'faith' in w]
faith_words = set(faith_words)
sorted(faith_words)
```

```python
bgrms = nltk.FreqDist(nltk.bigrams(bible_filtered2))
bgrms.most_common(20)
```

```python
tgrms = nltk.FreqDist(nltk.trigrams (bible_filtered2))
tgrms.most_common(20)
```

```python
#To check the number of times a word appeared in a text
occurences = bible_filtered2.count("stephen")
print('Number of occurences of word :', occurences)
```
